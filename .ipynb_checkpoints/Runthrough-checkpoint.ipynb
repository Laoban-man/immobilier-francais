```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
import glob
from dask import dataframe as dd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pickle 
from datetime import datetime
```

```python
file_path=os.path.join("..","data","valeurs-foncieres")
file_list=glob.glob(file_path+"/*.txt")
file_list
```

```python
data_main=pd.read_csv(file_list[0],sep="|",decimal=",")
data_main=data_main[data_main["Code postal"]>=75000]
data_main=data_main[data_main["Code postal"]<76000]
data_main["Surface_carrez_totale"]=data_main[["Surface Carrez du 1er lot","Surface Carrez du 2eme lot","Surface Carrez du 3eme lot","Surface Carrez du 4eme lot","Surface Carrez du 5eme lot"]].sum(axis=1,skipna=True)
for a in tqdm(range(1,len(file_list))):
    temp=pd.read_csv(file_list[a],sep="|",decimal=",")
    temp=temp[temp["Code postal"]>=75000]
    temp=temp[temp["Code postal"]<76000]
    temp["Surface_carrez_totale"]=temp[["Surface Carrez du 1er lot","Surface Carrez du 2eme lot","Surface Carrez du 3eme lot","Surface Carrez du 4eme lot","Surface Carrez du 5eme lot"]].sum(axis=1,skipna=True)
    data_main=pd.concat([data_main,temp])
```

```python
with open("consolidated_parision_data_unclean.pkl","wb") as f:
    pickle.dump(data_main,f)
```

```python

    
data=data_main[['Date mutation', 'Nature mutation', 'Valeur fonciere', 'No voie', 'Code voie', 'Voie', 'Code commune','Code postal','Nombre de lots','Code type local','Surface_carrez_totale', 'Nombre pieces principales']]
data['Date mutation']=pd.to_numeric(pd.to_datetime(data['Date mutation'], format='%d/%m/%Y').dt.year,downcast="integer")
data["Code postal"]=pd.to_numeric(data["Code postal"],downcast="integer")
data=data[data["Date mutation"]==2019]

data=data[data["Valeur fonciere"]<20000000]
data=data[data["Surface_carrez_totale"]>9]
data["Valeur_m2"]=data["Valeur fonciere"]/data["Surface_carrez_totale"]
from pandas_profiling import ProfileReport
profile=ProfileReport(data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def pivot_count(df, rows, columns, calc_field):
    df_pivot = df.pivot_table(values=calc_field, 
                              index=rows, 
                              columns=columns, 
                              aggfunc='mean'
                             ).dropna(axis=0)
    return df_pivot
data['binned_surface'] = pd.cut(data['Surface_carrez_totale'],bins=100)
effic_size_pivot = pivot_count(data,'binned_surface',
                               'Code postal','Valeur_m2')
plt.figure(figsize=(20,10))
sns.heatmap(effic_size_pivot, annot=True, fmt='g')
plt.title('Valeur au m2')
plt.ylabel('Surface')
plt.savefig("surfacexprix.png")
```

```python
data=pd.concat([data,pd.get_dummies(data["Nature mutation"],prefix="nature_mutation_")],axis=1)
data=pd.concat([data,pd.get_dummies(data["Code commune"],prefix="code_commune_")],axis=1)
data=pd.concat([data,pd.get_dummies(data["Code type local"],prefix="type_local_")],axis=1)
#data=pd.concat([data,pd.get_dummies(data["Code voie"],prefix="code_voie_")])
data.drop(["Date mutation","Nature mutation","Code commune","Code type local",'Code voie',"No voie","Voie","Code postal","Valeur fonciere"],axis=1,inplace=True)
data=data[data["Surface_carrez_totale"]>0]
#data.info()
data.head()
```

```python
from sklearn.impute import KNNImputer
from sklearn.impute import SimpleImputer


y=data["Valeur_m2"]
X=data.drop(["Valeur_m2"],axis=1)

impy = SimpleImputer(missing_values=np.nan, strategy='mean')
y=np.array(y).reshape(len(y),1)
y = impy.fit_transform(y)
y=pd.DataFrame(y,index=X.index)

impx = SimpleImputer(missing_values=np.nan, strategy='mean')
X = pd.DataFrame(impx.fit_transform(X),columns=X.columns,index=X.index)

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Scale the input features
#xscaler = StandardScaler()
#X_num_train=X_train[["Nombre de lots","Surface_carrez_totale","Nombre pieces principales"]]
#X_num_test=X_test[["Nombre de lots","Surface_carrez_totale","Nombre pieces principales"]]
#X_num_train = pd.DataFrame(xscaler.fit_transform(X_num_train),columns=X_num_train.columns,index=X_num_train.index)
#X_train["Nombre de lots"]=X_num_train["Nombre de lots"]
#X_train["Surface_carrez_totale"]=X_num_train["Surface_carrez_totale"]
#X_train["Nombre pieces principales"]=X_num_train["Nombre pieces principales"]
#X_num_test = pd.DataFrame(xscaler.transform(X_num_test),columns=X_num_test.columns,index=X_num_test.index)
#X_test["Nombre de lots"]=X_num_test["Nombre de lots"]
#X_test["Surface_carrez_totale"]=X_num_test["Surface_carrez_totale"]
#X_test["Nombre pieces principales"]=X_num_test["Nombre pieces principales"]

X_train=X_train[["Nombre de lots","Surface_carrez_totale","Nombre pieces principales"]]
X_test=X_test[["Nombre de lots","Surface_carrez_totale","Nombre pieces principales"]]
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import ModelCheckpoint
import keras
# Define a function

def model(input_dim):

    tf.keras.backend.clear_session()
    tf.random.set_seed(60)
    model=Sequential()
    model.add(Dense(500, input_dim=input_dim, activation= "relu"))
    model.add(Dense(100, activation= "relu"))
    model.add(Dense(50, activation= "relu"))
    model.add(Dense(1,activation="linear"))
    
    return model


my_model = model(input_dim=X_train.shape[1])

my_model.summary()
optimizer = keras.optimizers.Adam(lr=0.1)

checkpoint_name = 'Weights\Weights-{epoch:03d}--{val_loss:.5f}.hdf5' 
checkpoint = ModelCheckpoint(checkpoint_name, monitor='val_loss', verbose = 0, save_best_only = True, mode ='auto')
callbacks_list = [checkpoint]

my_model.compile(loss="mae" , optimizer=optimizer, metrics=['mse','mae'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Train the model, iterating on the data in batches of 32 samples
history =my_model.fit(X_train, y_train,
                    epochs=20, batch_size=1024,
                    validation_data=(X_test, y_test), 
                    callbacks=callbacks_list,
                    verbose=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error as mse

lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred_tr=lr.predict(X_train)
print(r2_score(y_train,y_pred_tr))
print(mse(y_train,y_pred_tr)**0.5)
y_pred_te=lr.predict(X_test)
print(r2_score(y_test,y_pred_te))
print(mse(y_test,y_pred_te)**0.5)
```

```python
import folium
import shapely.affinity
from shapely.geometry import Point
import geopy
import geopy.distance

# create an ellipse
radius_lat = geopy.distance.distance(kilometers = 10).destination(point=center, bearing=0)[0] - center[0]
radius_lon = geopy.distance.distance(kilometers = 10).destination(point=center, bearing=90)[1] - center[1]

circle = Point(center[::-1]).buffer(1)
ellipse = shapely.affinity.scale(circle, radius_lon, radius_lat)

ellipse_map = folium.Map(location=center, zoom_start=10, tiles='CartoDB positron')

folium.PolyLine(list([c[::-1] for c in ellipse.exterior.coords])).add_to(ellipse_map)

ellipse_map
```

```python
import math
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap


def plot_and_save(data,top,bottom,left,right,linewidth,legend_title,legend_bbox,path,colors,dpi=300):
    f, ax = plt.subplots(1, figsize=(15, 15))

    data.plot(ax=ax,color=[colors[str(t)] for t in data.prices_category],linewidth=linewidth)
    ax.set_aspect(1/math.cos(math.pi/180*center[0]))
    ax.set_ylim((bottom, top))
    ax.set_xlim((left, right))
    #f.colorbar(newcmp)
    plt.legend(handles=[mpatches.Patch(color=colors[str(k)],label=k) for k in set(paris["prices_category"])],loc='center right',fontsize='large',edgecolor='none',
                                   title=legend_title,title_fontsize='xx-large',bbox_to_anchor=legend_bbox)
    plt.savefig(path, bbox_inches='tight', dpi=dpi)
```

```python
plot_and_save(paris,top,bottom,left,right,0.7,'Paris',(1.25, 0.5),'parisstreetprices.png',colors)
```
